```python
l = [1,2,3,4,"sudh" , "kumar" , 324,34.456,"abc"]
```

```python
l1_num = []
l2_str = []
for i in l :
    if type(i) == int or type(i) == float :
        l1_num.append(i)
    else :
        l2_str.append(i)
```

```python
x = int(input("Enter a number :"))
temp = x
reverse = 0
while x > 0:
  digit = x%10
  reverse = (reverse * 10 )+digit
  x = x/10
if temp == reverse:
  print("The temp is palindrome!")
else:
  print("Not a palindrome!")
```
